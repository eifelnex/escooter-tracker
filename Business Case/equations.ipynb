# Business Case Model Equations

This notebook documents the mathematical formulation used to calculate weekly operating profit for e-scooter fleet operations.

## Core Equations

$$
\begin{aligned}
    R_w &= \sum_{\substack{\text{rides } i \\ \text{in week } w}} 
    f_p^{\text{unlock}} + t_i \cdot r_p
    & \text{(Revenue)} \\[6pt]
    %
    E_w &= \sum_{\substack{\text{idle drain} \\ \text{events } j \text{ in } w}} 
    \frac{\Delta d_j \cdot \varepsilon}{\eta_{\text{chg}}}
    + \sum_{\substack{\text{rides } i \\ \text{in week } w}} 
    \frac{d_i \cdot \varepsilon}{\eta_{\text{chg}}}
    & \text{(Energy)} \\[6pt]
    %
    C_w^{\text{elec}} &= E_w \cdot p_{\text{elec}}
    & \text{(Electricity Cost)} \\[6pt]
    %
    C_w^{\text{labor}} &= \frac{E_w}{\bar{E}_{\text{recharge}}} 
    \cdot t_{\text{recharge}} \cdot p_{\text{labor}}
    + n_{\text{repair}} \cdot t_{\text{repair}} \cdot p_{\text{labor}}
    & \text{(Labour Cost)} \\[6pt]
    %
    C_w^{\text{depr}} &= \sum_{\substack{\text{rides } i \\ \text{in week } w}} 
    d_i \cdot \frac{C_{\text{s}}}{L_{\text{km}}}
    + N_{\text{fleet}} \cdot \frac{C_{\text{s}}}{L_{\text{weeks}}}
    & \text{(Depreciation)} \\[8pt]
    %
    \Pi_w &= R_w - \bigl( C_w^{\text{elec}} + C_w^{\text{labor}} 
    + C_w^{\text{depr}} \bigr)
    & \text{(Profit)}
\end{aligned}
$$

## Symbol Reference

### Aggregate Quantities (per week $w$)

| Symbol | Description |
|:-------|:------------|
| $R_w$ | Total net revenue in week $w$ |
| $E_w$ | Total grid energy required for recharging in week $w$ (Wh) |
| $C_w^{\text{elec}}$ | Total electricity cost in week $w$ |
| $C_w^{\text{labor}}$ | Total labour cost in week $w$ |
| $C_w^{\text{depr}}$ | Total depreciation cost in week $w$ |
| $\Pi_w$ | Weekly operating profit |

### Revenue Parameters

| Symbol | Description |
|:-------|:------------|
| $f_p^{\text{unlock}}$ | Unlock fee charged by provider $p$ |
| $t_i$ | Duration of ride $i$ (min) |
| $r_p$ | Per-minute price of provider $p$ |

### Energy Parameters

| Symbol | Description |
|:-------|:------------|
| $d_i$ | Distance consumed during ride $i$ (km) |
| $\Delta d_j$ | Range lost during idle drain event $j$ (km) |
| $\varepsilon$ | Specific energy consumption (Wh/km) |
| $\eta_{\text{chg}}$ | Charger efficiency (grid-to-battery) |
| $p_{\text{elec}}$ | Electricity price (EUR/Wh) |

### Labour Parameters

| Symbol | Description |
|:-------|:------------|
| $\bar{E}_{\text{recharge}}$ | Average energy replenished per recharge event (Wh) |
| $t_{\text{recharge}}$ | Labour time per battery recharge (h) |
| $p_{\text{labor}}$ | Hourly labour rate (EUR/h) |
| $n_{\text{repair}}$ | Number of repairs per week |
| $t_{\text{repair}}$ | Labour time per repair (h) |

### Depreciation Parameters

| Symbol | Description |
|:-------|:------------|
| $C_{\text{s}}$ | Acquisition cost per scooter (EUR) |
| $L_{\text{km}}$ | Expected vehicle lifetime (km) |
| $L_{\text{weeks}}$ | Expected vehicle lifetime (weeks) |
| $N_{\text{fleet}}$ | Total fleet size (number of scooters) |
